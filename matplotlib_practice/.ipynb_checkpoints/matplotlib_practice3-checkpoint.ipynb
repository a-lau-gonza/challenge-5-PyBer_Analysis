```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.scatter(x_axis, y_axis, color="red", label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.legend()
plt.title('PyBer Fare by Month')
```

```python
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger, color="blue", label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.legend()
plt.title('PyBer Fare by Month')
```

```python
plt.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis], color="blue", label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.legend()
plt.title('PyBer Fare by Month')
```

```python
# Now we try the ObjOr technique:
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```
